```python
#import all required libraries
import pandas as pd
import yfinance as yf
import matplotlib as mp

#Testing as PoC, then using def for ease of use
```

```python
#Testing with Tesla Stock First
Close = yf.download("AAPL")
#AAPL_Close = Close.tail(15) #14 day RSI
AAPL_Close = Close["Close"]
AAPL_Close
```

```python
diff = AAPL_Close.diff()

gain = diff.clip(lower=0)
loss = diff.clip(upper=0)

avg_gain = gain.mean()
avg_loss = -loss.mean()

RS = avg_gain/avg_loss
RSI = 100 - (100/(1+RS))
print(RSI)
```

```python
AAPL_Close.plot()
```

```python
average_gain = gain.rolling(window=14).mean()
average_loss = loss.rolling(window=14).mean()
average_gain
average_loss

RS = average_gain/abs(average_loss)
RSI_test = 100 - (100/(1+RS))
print(RSI_test)
```

```python
RSI_test.plot()
```
